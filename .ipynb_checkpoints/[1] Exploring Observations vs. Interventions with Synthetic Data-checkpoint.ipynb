```python
# Below is a beginner-friendly exercise 
# in Python that illustrates the difference between observing a relationship between two variables 
# and intervening on one variable (i.e., the concept of do(X=x)). 
# While we won’t fully implement Pearl’s do-calculus, 
# this exercise provides intuition around why “doing” is different from merely “seeing.”
```

```python
# Import libraries

import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

sns.set(style = 'whitegrid') # for nicer plots
```

```python
# Three variables are considered:
# C (confounder); X (Treatment); Y (Outcome)
# We will follow Data-Generating Process, in which:
# 1. C ~ Uniform(0,10) (unifrom distribution: all values between 0 and 10 are equally likely)
# 2. X = 2C + εx (X depends on C (for example, C is 'lifestyle' and X is 'Physical Activity'))
# 3. Y = 3C + 1.5X + εy (Y depends on both C and X (Y might be 'Health Score'))
# εx and εy are random noise.
```

```python
# Generate Synthetic Data with a Confounder

np.random.seed(42) # for reproducibility

N = 200 # number of samples

# 1. confounder C
C = np.random.uniform(0, 10, N)

# 2. X depends on C
epsilon_x = np.random.normal(0, 1, N) # The noise 'εx' comes from a normal distribution with mean 0 and standard deviation 1.
X = 2*C + epsilon_x # X is correlated with C

# 3. Y depends on both C and X
epsilon_y = np.random.normal(0, 2, N)
Y = 3*C + 1.5*X

# Put it into a dataframe
df_variables = pd.DataFrame({'C': C, 'X': X, 'Y': Y})
df_variables.head()
```

```python
# Q: Based on the DGP, do you expect X and Y to be correlated?
# A: Yes. Because Y expilicitly depends on X, a higher X should correspond to a higher Y. Morever, 
#     because Y and X share common driver C - whenever it is higher, both Y (3C + 1.5X) and X (3C) is higher.
```

```python
# Observational Analysis: Correlation and Scatter Plot

# 1. Calculate Correlation between X and Y from the observed data.
corr_observed = df_variables[['X', 'Y']].corr().iloc[0,1]
print (f"Observed correlation between X and Y: {corr_observed:.2f}")

# 2. Visualize with a scatter plot.
sns.scatterplot(x='X', y='Y', data=df_variables)
plt.title(f"Observed Data (Correlation = {corr_observed:.2f})")
plt.show
```

```python
# Q: Are we sure this relationship is causal or could confounder C be influencing both?
# A: We cannot be sure that this relationship is causual. Actually, it can be a result of spurios correlation. X and Y appear to move together, 
#    but the real driver is C, as in the example of the spurios correlation between nicotine-stained fingers (X) and lung cancer (Y),
#    which confound by smoking (C).
```

```python
# 5. Naively Condition on X (Observational)
# P(Y∣X=x)

# Bin X into categories
df_variables['X_bin'] = pd.cut(df_variables['X'], bins=5, labels=False)

# Compute average Y in each bin
obs_means = df_variables.groupby('X_bin')['Y'].mean()
print("Observational means of Y by X bin:")
print(obs_means)

# Although binning means losing some detailes from the continuous variable, it nay be useful for exploratory insight and visual clarity.
# It's a tool rather than requierment. When I move to formal, I'd revert to the continuous version of the variable.
```

```python
# This table tells us what Y typocally is for different ranges of X.
```

```python
# 6. 6. Simulate Interventions
# P(Y∣do(X=x))

# Beacuse we stimulate an interventions, or - in other words - force X, we break the link to C. 
# We will choose set of "intervened" X, and for each "do(X=x)" scenario, we forced X = x, rather than X = 2*C + noise. However, we keep the
# relationship Y-X and Y-C (since Y still depends on C and X).

def simulate_intervention(num_samples=200, do_x_value=5):
    # Step 1: generate C as usual
    C_int = np.random.uniform(0, 10, num_samples) # We have 200 random values, with similair likelihood to be chosen, between 0 and 10.

    # Step 2: set X = do_x_value (forcing X)
    X_int = np.full(num_samples, do_x_value)

    # Step 2: Y depends on both C and forces X
    epsilon_y_int = np.random.normal(0, 2, num_samples)
    Y_int = 3*C_int + 1.5*X_int + epsilon_y_int

    return pd.DataFrame({'C': C_int, 'X_forced': X_int, 'Y': Y_int})
```

```python
# Test with a single intervention do(x=5)
df_int_5 = simulate_intervention(num_samples=200, do_x_value=5)
df_int_5.head()
```

```python
# Compare the means of Y in forced scenarios vs. observational data:

intervention_values = [0, 5, 10, 15, 20]
results = []

for val in intervention_values:
    df_int = simulate_intervention(num_samples=1000, do_x_value=val)
    avg_y = df_int['Y'].mean()
    results.append((val, avg_y))

df_results = pd.DataFrame(results, columns=['do(X)', 'Mean Y'])
print("Mean Y under different forced X values:")
print(df_results)
```

```python
# "Forced" means of Y might differ from the naive observational means (we've decoupled X from its confounder C).
```

```python
# 7. Compare Observational vs. Interventional Plots

plt.figure(figsize=(10,5))

# Observational data
plt.scatter(df_variables['X'], df_variables['Y'], alpha=0.4, label='Observed data')

# Interventional data
df_int_5 = simulate_intervention(num_samples=200, do_x_value=5)
plt.scatter(df_int_5['X_forced'], df_int_5['Y'], alpha=0.7, label='do(X=5)')

plt.xlabel('X')
plt.ylabel('Y')
plt.title('Observational vs. intervational data (X forced to 5)')
plt.legend()
plt.show()
```

```python
# In the forced case, X is the same for everyone, but Y still varies because of C (and noise)
# In the observational plot, X and Y are co-vary, but we don't know if that's the truly effect of X on Y or that's confounded by C.
# In the intervational plot, by forcing X to a specific value, we broke the influence of C on X. We get differnet distribution of Y that better 
# reflects the effect of X.
```

```python
# Key lessons:
# P(Y|X=x) != P(Y|do(X=x)).
# do(X=x) is about imposing X = x.
```

```python
# Questions
# 1. Plot the effect of C on X and the effect of C on Y. Is C confounder?

fig, axs = plt.subplots(nrows=1, ncols=2, figsize=(10,5))

# Subplot 1: C vs X
axs[0].scatter(df_variables['C'], df_variables['X'], alpha=0.4, color='tomato')
axs[0].set_xlabel('C')
axs[0].set_ylabel('X')
axs[0].set_title('C vs X')

# Subplot 2: C vs Y
axs[1].scatter(df_variables['C'], df_variables['Y'], alpha=0.4, color='steelblue')
axs[1].set_xlabel('C')
axs[1].set_ylabel('Y')
axs[1].set_title('C vs Y')

plt.tight_layout()
plt.show()
```

```python
# Based on the plots, the C may be confounder, as both X and Y tend to increase when C increases. That means X and Y
# might appear correlated even if the direct patway from X to Y is weaker.
```
